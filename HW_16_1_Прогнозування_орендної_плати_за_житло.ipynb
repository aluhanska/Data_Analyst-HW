```python
# 1 Спочатку кодуємо категоріальні змінні, що мають відносно небагато різних значень (у нашому випадку <= 6)

type_dummies = pd.get_dummies(df_clean['Area Type'], prefix='type')
city_dummies = pd.get_dummies(df_clean['City'], prefix='city')
status_dummies = pd.get_dummies(df_clean['Furnishing Status'], prefix='status')
tenant_dummies = pd.get_dummies(df_clean['Tenant Preferred'], prefix='tenant')
point_of_dummies = pd.get_dummies(df_clean['Point of Contact'], prefix='point_of')

display(df_clean['Area Type'].head())                                 # перевіряємо вихідні дані (одні з)
display(type_dummies.head())                                          # перевіряємо правильність кодування (bool type)
display(type_dummies.head().astype(int))                              # перевіряємо правильність кодування (int type)
```

```python
df_clean = pd.concat([df_clean, type_dummies, city_dummies, status_dummies, tenant_dummies, point_of_dummies], axis=1)
```

```python
# 2 Далі додаємо нові ознаки, обчислені на основі наявних даних:

# ознака № 1 - загальна кількість поверхів
# (орендар звертає увагу на поверховість будинку під час вибору квартири під оренду в першу чергу з огляду на наявність ліфта у високоповерховому будинку та його відсутність у низькоповерховому)

df_clean[['Floor_level', 'Total_floors']] = df_clean.Floor.str.split(' out of ', expand=True)
df_clean[['Floor_level', 'Total_floors']].head()
```

```python
# ознака № 2 - поверх
# (поверховість квартири також має значення, наприклад, в середньому на високих поверхах кращий вигляд, а до 5 поверху можна відносно легко піднятися пішки)

df_clean['Floor_level'] = df_clean['Floor_level'].replace({
    'Ground': 0,
    'Upper Basement': -1,
    'Lower Basement': -2
})
df_clean['Floor_level'].head()
```

```python
cat_cols = df_clean.select_dtypes('object').columns
df_clean[cat_cols].nunique()
```

```python
df_clean[['Total_floors', 'Floor_level']].isnull().sum()
```

```python
df_clean[df_clean.isnull().any(axis=1)]
```

```python
df_clean['Floor_level'] = pd.to_numeric(df_clean['Floor_level'], errors='coerce')
df_clean['Total_floors'] = pd.to_numeric(df_clean['Total_floors'], errors='coerce')
```

```python
df_clean['Total_floors'].describe().round(2)
```

```python
df_clean['Total_floors'] = df_clean['Total_floors'].fillna(df_clean['Total_floors'].median())
```

```python
df_clean['Total_floors'].isnull().sum()
```

```python
df_clean['Total_floors'].describe().round(2)
```

```python
# ознака 3 - нормалізована позиція поверху
# (необхідно для подальшого кодування даної категоріальної змінної)

def floor_cat(row):
  ratio = row['Floor_level'] / row['Total_floors']
  if ratio <= 0.33:
    return "low"
  elif ratio <= 0.66:
    return "middle"
  else:
    return "hight"

df_clean['floor_category'] = df_clean.apply(floor_cat, axis=1)
df_clean['floor_category']
```

```python
floor_dummies = pd.get_dummies(df_clean['floor_category'], prefix='floor')
df_clean = pd.concat([df_clean, floor_dummies], axis=1)
```

```python
# обробляємо дату публікації оголошення з метою виявлення ознак, які можуть бути корисними для моделі
# та обираємо наступні ознаки, які можуть описувати сезонність та тижневі коливання активності користувачів сайту оголошень:
# - № 4 - квартал;
# - № 5 - місяць;
# - № 5 - день тижня;
# - № 6 - чи вихідний

df_clean['Posted On'] = pd.to_datetime(df_clean['Posted On'])

df_clean['year'] = df_clean['Posted On'].dt.year
df_clean['quarter'] = df_clean['Posted On'].dt.quarter
df_clean['month'] = df_clean['Posted On'].dt.month
df_clean['day'] = df_clean['Posted On'].dt.day
df_clean['week_of_year'] = df_clean['Posted On'].dt.isocalendar().week
df_clean['weekday'] = df_clean['Posted On'].dt.weekday                               # 0 - понеділок, 6 - неділя
df_clean['is_weekend'] = df_clean['Posted On'].dt.weekday.isin([5, 6]).astype(int)

df_clean[['year', 'quarter', 'month', 'day', 'week_of_year', 'weekday', 'is_weekend']].nunique()
```

```python
# 3
df_clean = df_clean.drop(['Posted On', 'Floor', 'Area Type', 'Area Locality', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact', 'Floor_level', 'Total_floors', 'floor_category', 'year', 'day', 'week_of_year'], axis=1)
df_clean.head()
```
